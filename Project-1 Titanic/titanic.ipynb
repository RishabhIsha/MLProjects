```python
data1.head(5)
```

```python
data1.drop('Cabin',axis=1,inplace=True)
```

```python
data1.head(5)
```

```python
data1.dropna(inplace=True)
```

```python
data1.isnull().sum()
```

```python
sex=pd.get_dummies(data1['Sex'],drop_first=True)
```

```python
embark=pd.get_dummies(data1['Embarked'],drop_first=True)
embark.head(5)
```

```python
vclas=pd.get_dummies(data1['Pclass'],drop_first=True)
vclas.head(5)
```

```python
data1=pd.concat([data1,sex,embark,vclas],axis=1)
```

```python
data1.head(5)
```

```python
data1.drop(['PassengerId','Name','Sex','Embarked','Ticket'],axis=1,inplace=True)
```

```python
data1.head(5)
```

```python
data1.drop(['Pclass'],axis=1,inplace=True)
```

```python
data1.head(5)
```

## Train data

```python
X=data1.drop("Survived",axis=1)
Y=data1["Survived"]
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report as cl
```

```python
cl(y_test,prediction)
```

```python
from sklearn.metrics import confusion_matrix as cm
```

```python
cm(y_test,prediction)
```

```python
from sklearn.metrics import accuracy_score as a_sc
```

```python
a_sc(y_test,prediction)
```
